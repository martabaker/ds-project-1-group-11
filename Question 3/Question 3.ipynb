```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
 # Create reference to CSV file
file_path = "../Data/cleaned_car_crash_data.csv"

# Import the CSV into pandas DataFrame
df = pd.read_csv(file_path)

# Display first 5 Rows
df.head()
```

```python

filtered_df = df[['day', 'hour']]

# Group by the 'Category' column
monDay = filtered_df.groupby('day')['hour'].value_counts()[2].reset_index()
monDay = monDay.sort_values(by= ['hour'], ascending= True)
monDay = monDay.set_index('hour')

tuesDay = filtered_df.groupby('day')['hour'].value_counts()[3].reset_index()
tuesDay = tuesDay.sort_values(by= ['hour'], ascending= True)
tuesDay = tuesDay.set_index('hour')

wedDay = filtered_df.groupby('day')['hour'].value_counts()[4].reset_index()
wedDay = wedDay.sort_values(by= ['hour'], ascending= True)
wedDay = wedDay.set_index('hour')

thurDay = filtered_df.groupby('day')['hour'].value_counts()[5].reset_index()
thurDay = thurDay.sort_values(by= ['hour'], ascending= True)
thurDay = thurDay.set_index('hour')

friDay = filtered_df.groupby('day')['hour'].value_counts()[6].reset_index()
friDay = friDay.sort_values(by= ['hour'], ascending= True)
friDay = friDay.set_index('hour')

satDay = filtered_df.groupby('day')['hour'].value_counts()[7].reset_index()
satDay = satDay.sort_values(by= ['hour'], ascending= True)
satDay = satDay.set_index('hour')

sunDay = filtered_df.groupby('day')['hour'].value_counts()[1].reset_index()
sunDay = sunDay.sort_values(by= ['hour'], ascending= True)
sunDay = sunDay.set_index('hour')
```

```python
#Create the Line Graph
x = ['0', '100','200','300','400', '500','600','700','800','900','1000','1100','1200','1300','1400','1500','1600','1700','1800','1900','2000','2100','2200','2300']
fig, ax = plt.subplots()
plt.figure().set_figwidth(15)
plt.style.use('fivethirtyeight')
plt.plot(x,monDay, color='blue', label='Monday')
plt.plot(x,tuesDay, color='#0837a9', label='Tuesday')
plt.plot(x,wedDay, color='#0d7b07', label='Wednesday')
plt.plot(x,thurDay, color='#10b706', label='Thursday')
plt.plot(x,friDay, color='#00c6ff', label='Friday')
plt.plot(x,satDay, color='#08d77d', label='Saturday')
plt.plot(x,sunDay, color='#0f26df', label='Sunday')
plt.title('Total Number of Accidents by the Hour', fontsize=15)
plt.legend(title="Day",loc=2, fontsize='small', fancybox=True)
plt.ylabel('Number of Accidents', fontsize=13)
#plt.show()
```
